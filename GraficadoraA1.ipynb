```python
#Importamos las librerias que vamos a utilizar
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn')

#Este comando nos va a permitir graficar las funciones
%matplotlib notebook
```

```python
#Defino el limite que tendrá mi plano cartesiano
x = np.linspace(-10,10,1000)

#Defino las funciones a graficar
def f(x):
    return x**2-x+1
def g(x):
    return (2)/(x-1)
```

```python
#Este comando permite implantar las funciones en el plano cartesiano y mostrarlas
fig, ax = plt.subplots()
ax.plot(x,f(x))
ax.plot(x,g(x))
```
